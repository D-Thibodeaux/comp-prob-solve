```python
import numpy as np
import matplotlib.pyplot as plt
from homework_5_1_functions import random_sampling, importance_sampling
```

```python
#plot overlap integrals as a function of number of samples

#define figure
plt.figure(figsize=(8, 6))

random_overlap = []
importance_overlap = []

#calculate overlap integrals for different numbers of samples
for power in range(2,8):
    random_overlap.append([10 ** power, random_sampling(n_samples = 10 ** power)])
    importance_overlap.append([10 ** power, importance_sampling(n_samples = 10 ** power)])

#plot overlap integrals
plt.plot([x[0] for x in random_overlap], [x[1] for x in random_overlap], label = 'random sampling')
plt.plot([x[0] for x in importance_overlap], [x[1] for x in importance_overlap], label = 'importance sampling')

#format plot
plt.xscale('log')
plt.xlabel('Points Sampled')
plt.ylabel('Overlap Integral Value')
plt.title('Overlap Integral of Two H 2pz Orbitals')
plt.legend()
plt.show()
```

```python
"""
for both sampling methods, the value of the overlap integral converges with increasing sample size, and both methods converge to around the same value. 
for importance sampling the values of the overlap integral are closer to the value that it eventually converges to with lower sample size than with random sampling.
this is because the importance sampling biases the vaules at which the overlap integral is evaluated towards the regions that contribute more to the overlap integral. 
"""
```

```python
# plot overlap integral as a function of separation distance R

overlap = []
for r in [0.5 + 0.5 * x for x in range(40)]:
    #calculate and stor overlap integrals
    overlap.append([r, importance_sampling(n_samples = 10 ** 6, R = r)])

#plot overlap integral
plt.plot([x[0] for x in overlap], [x[1] for x in overlap])

#format plot
plt.xlabel('separation distance (a0)')
plt.ylabel('Overlap Integral Value')
plt.title('Overlap Integral of Two H 2pz Orbitals')
plt.show()
```

```python
"""
at R = 0, S(r) = 1 because the two H atoms are fully superimposed on eachother. 
As R increases, S(r) decreases and eventually becomes negative since the lobes of the orbitals between the two atoms have opposite phases, causing destructive interference.
as R goes to infinity, S(r) goes to 0 because the two orbitals are completely separate and no longer interact
"""
```
